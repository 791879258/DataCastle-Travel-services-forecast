```python
userid_rating_isnull=[]
for userid in history_zero_future_one:
    print(type(userComment_train[userComment_train['userid']==userid]['rating']))
    print(userComment_train[userComment_train['userid']==userid]['rating'].isnull().all())
    if(userComment_train[userComment_train['userid']==userid]['rating'].isnull().all()):
        userid_rate['nan'].append('nan')
        userid_rating_isnull.append(userid)
        continue
    else:
        userid_rate[userComment_train[userComment_train['userid']==userid]['rating'].values[0]].append(userid)
```
